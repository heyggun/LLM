**1. What is tokenization, and why is it important in LLMs?**

    - Tokenization is the process of splitting text into smaller units called tokens, which can be words, subwords, or even characters. 
    For instance, the word "tokenization" might be broken down into smaller subwords like "token" and "ization".
    This step is crucial because LLMs do not understand raw text directly. 
    Instead, they process sequences of numbers that represent these tokens.
    Effective tokenization allows models to handle various languages, manage rare words, and reduce the vocabulary size, which improves both efficiency and performance.

    - 토크나이제이션(Tokenization)은 텍스트를 토큰(Token) 이라는 작은 단위로 분할하는 과정으로,
    이 토큰은 단어, 부분 단어(subword), 혹은 문자(character) 단위 일 수 있다.
    예를 들어 "tokenizaion" 이라는 단어는 "token"과 "ization" 같은 작은 부분 단어로 나뉠 수 있다.
    이 과정은 매우 중요한데, LLM은 원시 텍스트(raw text)를 직접 이해하지 못하기 때문이다.
    대신, 모델은 이러한 토큰을 숫자로 변환한 시퀀스를 처리한다.
    효과적인 토크나이제이션은 모델이 여러 언어를 다룰 수 있또록 하며, 희귀 단어 처리를 용이하게 하고,
    어휘 크기를 줄여 연산 효율성과 성능을 동시에 개선하는 역할을 한다.



**2.What is LoRA and QLoRA?**

    LoRA and QLoRA are techniques designed to optimize the fine=tuning of Large Language Models (LLMs), focusing on reducing memory usage and enhancing efficiency without compromising performance in Natural Language Processing (NLP) tasks.



`LoRA(Low-Rank Adaptation)`

    LoRA is a parameter-efficient fine-tuning method that introduces new trainable parameters to modify a model's behavior without increasing its overall size.
    By doing so, LoRA maintains the original parameter count, reducing the memory overhead typically associated with training large models. It works by adding low-rank matrix adaptations to the model's existing layers, allowing for significant performance improvements while keeping resource consumption in check.
    This makes it ideal for environments where computational resources are limited, yet high model accuracy is still required.
    



`QLoRA (Quantized LoRA)`
   
    QLoRA builds on LoRA by incorporating quantization to further optimize memory usage. It uses techniques such as 4-bit Normal Float, Double Quantization and Paged Optimizers to compress the model's parameters and improve computational efficiency. By reducing the precision of model weights (e.g., from 16-bit to 4-bit) while retaining most of the model's accuracy, QLoRA allows for the fine-tuning of LLMs with minimal memory footprint. This method is particularly useful when scaling large models, as it maintains performance levels comparable to full-precision models while significantly reducing resource consumption.


    - LoRA와 QLoRA는 대형 언어 모델(LLM)의 파인튜닝을 최적화하기 위한 기술로, 자연어 처리(NLP) 작업에서 성능을 유지하면서도 메모리 사용량을 줄이고 효율성을 높이는데 초점을 맞춘다.

    `LoRA(Low-Rank Adaptation)`
    - LoRA는 모델의 전체 크기를 증가시키지 않으면서 새로운 학습 가능한 파라미터를 추가하여 모델의 동작을 조정하는 파라미터 효율적(Parameter-Efficient) 파인튜닝 기법이다. 기존 모델의 총 파라미터 수를 유지하면서도, 대형 모델을 학습할 때 발생하는 메모리 부담을 줄이는 효과가 있다. 이를 위해, 모델의 기존 레이어에 저차원 행렬(low-rank matrix) 기반의 적응(adaptation) 구조를 추가하여 모델의 성능을 크게 향상시키면서도 연산 및 메모리 자원을 효율적으로 활용할 수 있도록 한다. 이러한 특성 덕분에 LoRA는 연산 자원이 제한적이지만 높은 모델 정확도가 필요한 환경에서 특히 유용하다.
    
    `QLoRA(Quantized LoRA)`
    - QLoRA는 LoRA에 양자화(Quantization) 기법을 추가하여 메모리 사용을 더욱 최적화한 방식이다.
    4-bit Normal Float, Doubole Quantization, Paged Optimizers 등의 기술을 활용해 모델의 파라미터를 압축하고 연산 효율성을 개선한다. 모델 가중치의 표현 정밀도를 줄이면서(예: 16비트 -> 4비트)도 원래 모델의 성능을 대부분 유지할 수 있또록 설계 되었다. 이를 통해 LLM의 파인튜닝을 최소한의 메모리로 수행할 수 있으며, 특히 대형 모델을 확장할 때 정밀도를 유지하면서도 자원 소모를 획기적으로 줄일 수 있는 장점이 있다. 결과적으로 QLoRA는 풀 프리시전(full-precision) 모델과 유사한 성능을 유지하면서도 훨씬 적은 연산 자원으로 동작할 수 있어, 대규모 모델을 효율적으로 다룰 수 있는 강력한 기법이다. 


**3. What is beam search, and how doese it differ from greedy decoding?**

    Beam search is a search algorithm used during text generation to find the most likely sequence of words. Instead of choosing the single highest-probability word at each step(as greedy decoding does), beam search explores multiple possible sequences in parallel, maintaining a set of the top k candidates (beams). It balances between finding high-probability sequences and exploring alternative paths. This leads to more coherent and contextually appropriate outputs, especially in long-term text generation tasks.




    - 빔 서치(Beam Search)는 텍스트 생성 과정에서 가장 가능성이 높은 단어 시퀀스를 찾기 위해 사용되는 탐색 알고리즘이다.
    탐욕적 디코딩(Greedy Decoding) 처럼 매 단계에서 단일 최고 확률 단어만 선택하는 것이 아니라, 여러 개의 후보 시퀀스를 병렬로 유지하면서 탐색한다.
    이때, 상위 k개의 후보(beam)을 유지하며, 확률이 높은 경로를 따르면서도 다양한 가능성을 탐색할 수 있도록 균형을 조정한다.
    특히 긴 텍스트를 생성할 때 더 일관되고 문맥적으로 적절한 결과를 생성하는데 효과적이다.

**4. Explain the concept of temperature in LLM text generation**

    - Temperature is a hyper parameter that controls the randomness of text generation by adjusting the probability distribution over possible next tokens. A low temperature (close to o) makes the model highly deterministic, favoring the most probable tokens. 
    Conversely, a high temperature (above 1) encourages more diversity by flattening the distribution, allowing less probable tokens to be selected. For instance, a temperature of 0.7 strikes a balance between` creativity and coherence, making it suitable for generating diverse but sensible outputs.




    - Temperature는 텍스트 생성 시 출력의 무작위성을 조절하는 하이퍼파라미터로, 다음 토큰을 선택할 때의 확률 분포를 조정하는 역할을 한다. 
    낮은 Temperature(0에 가까울수록) -> 모델이 결정론적(deterministic)으로 동작하며, 가장 높은 확률을 가진 토큰을 선택하는 경향이 강하다. 이로 인해 일관성은 높아지지만, 창의성이 제한될 수 있다.
    높은 Temperature(1 이상) -> 확률 분포를 평탄하게 만들어, 상대적으로 낮은 확률을 가진 토큰도 선택될 가능성이 높아진다. 다양한 출력을 생성할 수 있지만 문맥적 일관성이 낮아질 수 있다.
    Temperature 0.7 정도는 창의성과 일관성의 균형을 맞추는 설정으로, 다양하면서도 말이 되는 출력을 생성하는데 적합하다.


**5. What is masked language modeling, and how does it contribute to model pretraining?**

    Masked language modeling (MLM) is a training objective where some tokens in the input are randomly masked, and the model is tasked with predicting them based on context. This forces the model to learn contextual relationships between words, enhancing its ability to understand language semantics. MLM is commonly used in models like BERT, which are pretrained using this objective to develop a deep understanding of language before fine-tuning on specific tasks.



    마스킹 언어 모델링(MLM, Maksed Language Modeling)은 입력  텍스트에서 일부 토큰을 무작위로 마스킹(masking) 하고, 모델이 주어진 문맥을 기반으로 해당 토큰을 예측하도록 학습하는 방식이다.
    이 과정은 단어 간의 문맥적 관계를 학습하도록 강제하여, 모델이 언어의 의미적 구조를 더 깊이 이해할 수 있도록 돕는다.
    MLM은 대표적으로 BERT와 같은 사전 학습 모델(pretrained model)에서 사용되며, 특정 태스크에 맞춰 파인 튜닝(fine-tuning) 되기 전에 언어에 대한 깊은 이해를 형성하는 핵심 학습 기법이다. 이러한 목표로 사전 학습을 거친 후 특정 작업에 맞게  파인 튜닝 된다.
    단순한 다음 단어 예측과 달리, 양방향(bidirectional) 문맥을 활용해 마스킹된 단어를 복원하는 방식이므로,
    보다 강력한 언어 표현력을 갖춘 모델을 만들 수 있다. 

**6. What are Sequence-to-Sequence Models?**

    Sequence-to-Sequence (Seq2Seq) models are a type of neural network architecture designed to transform one sequences of data into another sequence. These models are commonly used in tasks where the input and output have variable lengths, such as in machine translation, text summarization, and speech recognition.



    - 시퀀스-투-시퀀스(Sequence-to-Sequence) 모델은 하나의 데이터 시퀀스를 다른 시퀀스로 변환하는 신경망 아키텍처이다.
    이 모델은 입력과 출력의 길이가 가변적인 작업에서 자주 사용되며, 대표적으로 기계 번역, 텍스트 요약, 음성 인식 등의 작업에 활용된다.

**7. How do autoregressive models differ from masked models in LLM training?**

    Autogressive models, such as GPT, generate text on token at a time, with each token predicted based no the previously generated tokens. This sequential approach is ideal for tasks like text generation. Masked models, like BERT, predict randomly masked tokens within a sentence, leveraging both left and right context. Autoregressive models excel in generative tasks, while masked models are better suited for understanding and classification tasks.



    - 자기회귀 모델(Autoregressive models), 예를 들어 GPT는 텍스트를 한 번에 하나의 토큰씩 생성하며, 각 토큰은 이전에 생성된 토큰들을 바탕으로 예측된다. 이 순차적 접근법은 텍스트 생성과 같은 작업에 이상적이다.
    마스킹 모델(Masked models), 예를 들어 BERT는 문장 내에서 무작위로 마스킹된 토큰을 예측하며, 좌측과 우측 문맥을 모두 활용한다.
    자기회귀 모델은 생성 작업에서 뛰어난 성능을 보이며, 마스킹 모델은 이해 및 분류 작업에 더 적합하다.


**8. What role do embeddings play in LLMs, and how are they initialized?**

    Embeddings are dense, continuous vector representations of tokens, capturing semantic and syntactic information. They map discrete tokens (words or subwords) into a high-dimensional space, making them suitable for input into neural networks. Embeddings are typically initialized randomly or with retrained vectors like Word2Vec or Glove. During training, these embeddings are fine-tuned to capture task-specific nuances, enhancing the model's performance on various language tasks.



    - 임베딩(Embedding)은 토큰의 밀집된 연속 벡터 포현으로, 의미적 및 구분적 정보를 담고 있다.
    임베딩은 불연속적인 토큰(딴어 또는 부분 단어)을 고차원 공간에 매핑하여, 신경망에 입력하기 적합하게 만든다.
    임베딩은 일반적으로 무작위로 초기화되거나 Word2Vec이나 Glove와 같은 재학습된 벡터로 초기화된다.
    학습과정에서 이러한 임베딩은 작업별 세부 사항(task-specific nuances)을 반영하도록 미세조정(fine-tuning) 되어,
    모델의 다양한 언어 작업에 대한 성능을 향상시킨다.

**9. What is next sentence prediction and how is useful in language modelling?**

    Next Sentence Prediction (NSP) is a key technique used in language modeling, particularly in training large models like BERT(Bidirectional Encoder Representations from Transformers). NSP helps a model understand the relationship between two sentences, generation, and information retrieval.



    During pre-training, the model is fed two sentences:
    - 50% of the time, the second sentence is the actual next sentence in the document (positive pairs).
    - 50% of the time, the second sentence is a random sentence from the corpus (negative Paris). The model is trained to classify whether the second sentence is the correct next sentence sentence for not. This binary classification task is used alongside a masked language modeling task to improve the model's overall language understanding.


    다음 문장 예측(Next Sentence Prediction, NSP)는 언어 모델링에서 중요한 기법으로, 특히 BERT(Bidirectional Encoder Representations from Transformers)와 같은 대형 모델을 훈련할 때 사용된다.
    NSP는 모델이 두 문장 간의 관계를 이해하도록 도와주며, 생성 및 정보 검색 작업에 유용하다.
    - 50%의 경우, 두 번째 문장은 문서에서 실제로 다음 문장(끙정 쌍)이다.
    - 50%의 경우, 두 번째 문장은 코퍼스에서 임의로 선택된 문장(부정 쌍)이다.
    모델은 두 번째 문장이 올바른 다음 문장인지 아닌지를 이진 분류하는 방식으로 학습된다.
    이 이진 분류 작업은 마스킹 언어 모델링(Masked Language Modeling, MLM) 작업과 함께 사용되어, 모델의 전반적인 언어 이해 능력을 향상시킨다. 

**10. Explain the difference between top-k sampling and nuclues (top-p) sampling in LLMs.**

    Top-k sampling restricts the model's choices to the top k most probable tokens at each step, introducing controlled randomness. For example, setting k=10 means the model will only consider the 10 most likely tokens. Nucleus sampling, or top-p sampling, takes a more dynamic approach by selecting tokens whose cumulative probability exceeds a threshold p (e.g., 0.9). This allows for flexible candidate sets based on context, promoting both diversity and coherence in generated text.



    - `Top-k 샘플링(Top-k sampling)`은 각 단계에서 모델의 선택을 가장 확률이 높은 상위 k개의 토큰으로 제한하여 제어된 무작위성을 도입하는 방식이다. 예를 들어, k=10으로 설정하면 모델은 상위 10개의 가장 가능성이 높은 토큰만 고려하게 된다.

    -`뉴클레어스 샘플링(Nucleus sampling)` 또는 `Top-p`` 샘플링은 보다 동적인 접근법을 취한다.
    이 방식은 누적 확률이 p(예: 0.9)를 초과하는 토큰들을 선택한다. 이렇게 하면 문맥에 따라 유연한 후보 집합을 형성할 수 있어, 생성된 텍스트에서 다양성과 일관성을 동시에 얻을 수 있다.
